```python
import pandas as pd
```

```python
df = pd.read_csv('/opt/data/in_rolls/in_rolls_state_name_age_sex_fn_ln.csv.gz', usecols=['state', 'birth_year', 'first_name', 'sex'])
df
```

```python
df.dropna(subset=['first_name'], inplace=True)
df
```

* i think it makes sense to only release data where count for last_name/first_name is > 100
* it will take out a bunch of weird names for now

```python
adf = df[df['first_name'].groupby(df['first_name']).transform('size')>100]
```

```python
adf
```

```python
adf['first_name'].groupby(adf['first_name']).size()
```

```python
adf = adf[adf.first_name.str.contains('[a-z]')]
adf = adf[~adf.first_name.str.contains('\.')]
adf
```

```python
gdf = adf.groupby(['state', 'birth_year', 'first_name']).sex.value_counts().unstack(fill_value=0)
```

```python
gdf
```

```python
gdf.reset_index(inplace=True)
gdf
```

```python
gdf.columns = ['state', 'birth_year', 'first_name', 'n_female', 'n_male', 'n_third_gender']
gdf
```

```python
gdf['prop_female'] = gdf['n_female'] / (gdf['n_female'] + gdf['n_male'] + gdf['n_third_gender'])
gdf
```

```python
gdf.to_csv('/opt/data/in_rolls/in_rolls_state_year_fn_naampy.csv.gz', index=False, compression='gzip')
```

```python
bdf = gdf.groupby(['first_name']).agg({'n_female': 'sum', 'n_male': 'sum', 'n_third_gender': 'sum'}).reset_index()
bdf['prop_female'] = bdf['n_female'] / (bdf['n_female'] + bdf['n_male'] +  gdf['n_third_gender'])
bdf
```
