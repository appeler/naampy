```python
import pandas as pd
from glob import glob
```

```python
adf = None
```

```python
for fn in sorted(glob('*.tab')):
    state = fn.split('.')[0].split('-')[0]
    #df = pd.read_csv(fn, sep='\t', nrows=10, usecols=['elector_name', 'age', 'sex'])
    df = pd.read_csv(fn, sep='\t', usecols=['elector_name', 'age', 'sex'])
    df['state'] = state
    if adf is None:
        adf = df
    else:
        adf = pd.concat([adf, df])
    print(state, df.columns)
```

```python
adf
```

```python
for fn in sorted(glob('andhra/*.csv')):
    state = fn.split('.')[0].split('-')[0]
    #df = pd.read_csv(fn, nrows=10, usecols=['elector_name', 'age', 'sex'])
    df = pd.read_csv(fn, usecols=['elector_name', 'age', 'sex'])
    df['state'] = state.split('/')[0]
    if adf is None:
        adf = df
    else:
        adf = pd.concat([adf, df])
    print(state, df.columns)
```
