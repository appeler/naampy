```python
CHUNK_SIZE = 1000000
output_file = "/opt/data2/in_rolls/in_rolls_state_name_age_sex.csv.gz"
```

```python
sex_remap = {"महिला": "Female", "पुरूष": "Male", "स्त्री": "Female"}
```

```python
import time
```

```python
%%time
for fn in sorted(glob("/opt/data2/parse-in-rolls/clean+t13n/*.gz")):
    df = pd.read_csv(fn, nrows=1)
    cols = ["age", "sex", "state"]
    if "elector_name_t13n" in df.columns:
        cols.append("elector_name_t13n")
    else:
        cols.append("elector_name")
    count = 0
    total = 0
    for adf in pd.read_csv(fn, usecols=cols, chunksize=CHUNK_SIZE):
        t1 = time.time()
        adf.state = adf.state.str.lower()
        adf.rename(columns={"elector_name_t13n": "elector_name"}, inplace=True)
        adf.sex.replace(sex_remap, inplace=True)
        # print(adf.sex.unique())
        adf = adf[["elector_name", "age", "sex", "state"]]
        # print(fn, adf.columns)
        adf.to_csv(output_file, mode="a", header=False, index=False, compression="gzip")
        count += 1
        total += len(adf)
        t2 = time.time()
        print(count, total, t2 - t1, fn)
        # break
```
