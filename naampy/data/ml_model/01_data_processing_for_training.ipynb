```python
import pandas as pd
```

```python
df = pd.read_csv('in_rolls_state_year_fn_naampy_100.csv')
```

```python
df.head()
```

```python
agg_dict = {'n_female': 'sum', 'n_male': 'sum'}
```

```python
df = df.groupby(['first_name']).agg(agg_dict).reset_index()
```

```python
df.head()
```

```python
df['gender'] = ['female' if n_female > n_male else 'male' for n_female,n_male in zip(df['n_female'], df['n_male'])]
```

```python
df.head()
```

```python
df[df['first_name'] == 'amitabh']
```

```python
df['female_prop'] = [ (n_female/(n_male+n_female))  for n_female, n_male in zip(df['n_female'], df['n_male'])]
```

```python
df['male_prop'] = [ (n_male/(n_male+n_female))  for n_female, n_male in zip(df['n_female'], df['n_male'])]
```

```python
df.head()
```

```python
df[['first_name', 'gender', 'female_prop']].to_csv('naampy_final_prop.csv', index=False)
```
